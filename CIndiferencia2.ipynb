Para la función de utilidad que representa las preferencias racionales de la forma,
$$u(x,y) = \min\{x+2y,2x+y\}$$
vamos a esbozar la gráfica de curva de indiferencia correspondiente a nivel de utilidad $\overline{u} = 10$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de utilidad
def utilidad(x, y):
    return np.minimum(x + 2*y, 2*x + y)

# Valores para el nivel de utilidad constante
u0 = 10

# Crear un espacio de valores para x y y
x = np.linspace(0, 15, 400)
y = np.linspace(0, 15, 400)
X, Y = np.meshgrid(x, y)

# Calcular los valores de la función de utilidad
Z = utilidad(X,Y)

# Graficar la curva de indiferencia para un nivel de utilidad U0
plt.contour(X, Y, Z, levels=[u0], colors='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Curva de Indiferencia para u(x,y) = min(x+2y,2x+y) = {u0}')
plt.grid(True)
plt.show()
```

Donde las rectas del sistema de ecuaciones son:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones para las dos rectas
def recta1(x):
    return (10 - x) / 2  # Ecuación para x + 2y = 10

def recta2(x):
    return 10 - 2 * x    # Ecuación para 2x + y = 10

# Crear valores de x para graficar
x_vals = np.linspace(0, 10, 400)

# Calcular los valores de y para cada recta
y_vals_recta1 = recta1(x_vals)
y_vals_recta2 = recta2(x_vals)

# Calcular el punto de intersección
# x = 10/3 y y = 10/3
interseccion_x = 10/3
interseccion_y = 10/3

# Graficar las rectas y el punto de intersección
plt.plot(x_vals, y_vals_recta1, label=r'$x + 2y = 10$', color='blue')
plt.plot(x_vals, y_vals_recta2, label=r'$2x + y = 10$', color='green')
plt.scatter(interseccion_x, interseccion_y, color='red', zorder=5, label='Punto de intersección (10/3, 10/3)')



# Añadir etiquetas y leyenda
plt.title(r'Gráfica de las ecuaciones $x + 2y = 10$ y $2x + y = 10$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

# Mostrar la cuadrícula y gráfico
plt.grid(True)
plt.show()
```
