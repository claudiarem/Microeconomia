Para la función de utilidad que tiene la forma,
$$u(x,y) = x-y$$
Graficar algunas curvas de indiferencia.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de utilidad
def utilidad(x, y):
    return x - y

# Crear un espacio de valores para x e y
x = np.linspace(0, 20, 400)
y = np.linspace(0, 25, 400)
X, Y = np.meshgrid(x, y)

# Calcular los valores de la función de utilidad
Z = utilidad(X,Y)

# Valores para los dinstintos niveles de utilidad u0
niveles_utilidad = [2.5,5,10,15,20]

# Graficar la curva de indiferencia para un nivel de utilidad U0
plt.contour(X, Y, Z, levels=niveles_utilidad, colors='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Curva de Indiferencia para u(x,y) = x-y')
plt.grid(True)

# Añadir etiquetas a las curvas para indicar el nivel de utilidad
plt.clabel(plt.contour(X, Y, Z, levels=niveles_utilidad, colors='blue'), inline=True, fontsize=12)

plt.show()
```

Donde las rectas del sistema de ecuaciones son:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones para las dos rectas
def recta1(x):
    return 2*x - 10  # Ecuación para 2x - y = 10

def recta2(x):
    return (10 + x ) / 2    # Ecuación para 2y - x = 10

# Crear valores de x para graficar
x_vals1 = np.linspace(0, 20, 400)
x_vals2 = np.linspace(0, 20, 400)

# Calcular los valores de y para cada recta
x_vals1_recta1 = recta1(x_vals1)
x_vals2_recta2 = recta2(x_vals2)

# Graficar las rectas
plt.plot(x_vals1, x_vals1_recta1, label=r'$2x - y = 10$', color='blue')
plt.plot(x_vals2, x_vals2_recta2, label=r'$2y - x = 10$', color='green')

#Graficar el punto de intersección (x,y)= (10/3,10/3)
plt.plot(10, 10, 'ro', label='Intersección')

# Añadir etiquetas y leyenda
plt.xlim(0, 20) #limites del eje x
plt.ylim(0, 20) #limites del eje y
plt.title(r'Gráfica de las ecuaciones $2x - y = 10$ y $2y - x = 10$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

# Mostrar la cuadrícula y gráfico
plt.grid(True)
plt.show()
```
